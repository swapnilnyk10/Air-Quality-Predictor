In our dataset the missing values are marked with -200
Mentioned in UCI Documentation

```python
#Counting number of times -200 appears in dataframe
air_quality_data.isin([-200]).sum(axis=0)
```

We will replace all the missing values i.e. -200 with the mean of all the values present in the same column

For this first we will convert -200 to NaN and then compute the mean of non-NaN values then replace all NaN values with the computed mean value

```python
air_quality_data = air_quality_data.replace(to_replace = -200, value = np.NaN)
```

```python
#Converted -200 to NaN
#Now printing number of null values
air_quality_data.isnull().sum()
```

```python
air_quality_data.mean()
```

```python
#Replacing NaN values with mean values
air_quality_data = air_quality_data.fillna(air_quality_data.mean())
```

Currently in our dataset we have date format as DD/MM/YYYY but in Prophet model the supported format is YYYY-MM-DD for a date or YYYY-MM-DD HH:MM:SS for a timestamp.

```python
#Converting date format to YYYY-MM-DD
date_info = pd.to_datetime(air_quality_data['Date'])
date_info
```

```python
#Converting time format from HH.MM.SS to HH:MM:SS
time_info  = air_quality_data['Time']
time_info = time_info.apply(lambda x:x.replace('.', ':'))
time_info
```

```python
print(type(date_info))
print(type(time_info))
```

```python
#Combining two series to pandas dataframe
date_time = pd.concat([date_info, time_info], axis = 1)
```

```python
date_time.head()
```

Here we have two separate columns of date and time now we have to combine them in one column

```python
#Combining date and time
date_time['ds'] = date_time['Date'].astype(str) + ' ' + date_time['Time'].astype(str)
```

```python
date_time
```

```python
date_time.info()
```

Here the ds field is treated as an object as we converted it to a string now converting it to dateTime field

```python
data = pd.DataFrame()
data['ds'] = pd.to_datetime(date_time['ds'])
```

```python
data
```

```python
data['y'] = air_quality_data['RH']
```

```python
data
```

```python
from prophet import Prophet
```

```python
#Training the Prophet Model
model = Prophet(yearly_seasonality=True)
model.fit(data)
```

```python
#Making Future predictions
future = model.make_future_dataframe(periods=365, freq = 'H')
future
```

```python
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = model.plot(forecast)
```
